```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/Users/personal/Documents/Imarticus/Classroom/Python/EDA/Projects/Global Ai content impact/Global_AI_Content_Impact_Dataset.csv")

# Display first few rows
df.head()
```

## 1. Understand the Dataset

**Objective**: To explore the global impact of AI content across different countries and industries.

**Goals**:
- Analyze AI adoption, content generation, job impact, and revenue changes across sectors and countries.
- Identify trends, anomalies, and correlations in AI impact metrics.
- Understand regulation status and consumer trust levels in different regions.


```python
# Basic dataset information
df.info()

# Check for missing values
df.isnull().sum()

# Check for duplicates
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")

# Remove duplicates if any
df = df.drop_duplicates()

# Check data types
df.dtypes
```

## 3. Descriptive Statistics

```python
# Summary statistics
df.describe()
```

## 4. Univariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram for numeric columns
df.hist(figsize=(15, 10), bins=15)
plt.tight_layout()
plt.show()

# Bar plot for top industries
plt.figure(figsize=(10, 6))
sns.countplot(y='Industry', data=df, order=df['Industry'].value_counts().index)
plt.title("Industry Distribution")
plt.show()
```

## 5. Bivariate and Multivariate Analysis

```python
# Correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# Scatter plot: AI Adoption Rate vs Revenue Increase
sns.scatterplot(data=df, x='AI Adoption Rate (%)', y='Revenue Increase Due to AI (%)', hue='Industry')
plt.title("AI Adoption vs Revenue Increase")
plt.show()

# Boxplot of Job Loss by Regulation Status
plt.figure(figsize=(8, 6))
sns.boxplot(x='Regulation Status', y='Job Loss Due to AI (%)', data=df)
plt.title("Job Loss due to AI by Regulation Status")
plt.show()
```

## 6. Feature Engineering

```python
# Encode categorical variables
df_encoded = pd.get_dummies(df, columns=['Country', 'Industry', 'Top AI Tools Used', 'Regulation Status'])

# Show encoded columns
df_encoded.head()
```

## 7. Additional Visualization

```python
# Pairplot for key numeric variables
sns.pairplot(df[['AI Adoption Rate (%)', 'Job Loss Due to AI (%)',
                 'Revenue Increase Due to AI (%)', 'Consumer Trust in AI (%)']])
plt.show()
```

## 8. Hypothesis Testing

```python
from scipy.stats import ttest_ind

# Compare job loss in Strict vs Moderate regulation countries
strict = df[df['Regulation Status'] == 'Strict']['Job Loss Due to AI (%)']
moderate = df[df['Regulation Status'] == 'Moderate']['Job Loss Due to AI (%)']

t_stat, p_val = ttest_ind(strict, moderate)
print(f"T-Statistic: {t_stat:.4f}, P-Value: {p_val:.4f}")

if p_val < 0.05:
    print("Significant difference in job loss between Strict and Moderate regulation countries.")
else:
    print("No significant difference in job loss between Strict and Moderate regulation countries.")
```

## 9. Insights and Summary

- **AI Adoption** varies significantly across industries, with Automotive and Legal sectors showing high adoption.
- **Job Loss** tends to be higher in countries with strict regulations.
- **Revenue Increase** correlates moderately with AI Adoption Rate.
- **Consumer Trust** and **Human-AI Collaboration** are key metrics for evaluating AI's societal impact.
- There's statistical evidence suggesting regulation type may influence job loss due to AI.
